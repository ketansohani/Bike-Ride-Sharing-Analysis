### Outlier Detection and Treatment

```python
num_features=['temp','humidity','count']
```

```python
def z_score(data):
    count=0
    for i in data:
        z=np.abs((i-mean)/std)
        if z>3:
            count=count+1
    return count
```

```python
for feature in num_features:
    mean =  np.mean(df[feature])
    std = np.std(df[feature])
    outliers = z_score(df[feature])
    print("Number of Outliers in "+ feature + " = " + str(outliers))
```

So, we have 2 features with outliers present. We will be removing these outliers from our dataset.

```python
sns.boxplot(df['humidity'])
```

```python
sns.boxplot(df['count'])
```

```python
#Outlier removal for "Humidity"
percentile25 = df['humidity'].quantile(0.25)
percentile75 = df['humidity'].quantile(0.75)
iqr=percentile75-percentile25
upper_limit = percentile75 + (1.5 * iqr)
lower_limit = percentile25 - (1.5 * iqr)
df=df[(df['humidity']>lower_limit) & (df['humidity']<upper_limit)]
```

```python
sns.boxplot(df['humidity'])
```

```python
#Outlier removal for "Count"
percentile25 = df['count'].quantile(0.25)
percentile75 = df['count'].quantile(0.75)
iqr=percentile75-percentile25
upper_limit = percentile75 + (1.5 * iqr)
lower_limit = percentile25 - (1.5 * iqr)
df=df[(df['count']>lower_limit) & (df['count']<upper_limit)]
```

# Input Split

```python
X = df.drop(columns=['atemp', 'windspeed', 'casual', 'registered', 'count'], axis=1)
y = df['count']
```

```python
from sklearn.linear_model import LinearRegression, Ridge, HuberRegressor, ElasticNetCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor

models = [LinearRegression(),
         Ridge(),
         HuberRegressor(),
         ElasticNetCV(),
         DecisionTreeRegressor(),
         RandomForestRegressor(),
         ExtraTreesRegressor(),
         GradientBoostingRegressor()]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
def train(model, X, y):
    # train the model
    model.fit(X, y)
    
    # predict the training set
    pred = model.predict(X)
    
    # perform cross-validation
    cv_score = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("MSE:",mean_squared_error(y,pred))
    print("CV Score:", cv_score)
```

```python
model = DecisionTreeRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
coef.plot(kind='bar', title="Feature Importance",figsize=(10,8))
```

```python
from sklearn import model_selection
def train(model):
    kfold = model_selection.KFold(n_splits=5)
    pred = model_selection.cross_val_score(model, X, y, cv=kfold, scoring='neg_mean_squared_error')
    cv_score = pred.mean()
    print('Model:',model)
    print('CV score:', abs(cv_score))
```

```python
for model in models:
    train(model)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
param_dist = {"n_estimators":[100,200,300,500,700,1000],
              "max_features": [5,10,20,25,30,30],
              "bootstrap": [True, False],
                'criterion':['mae','mse'],
                'max_depth':[None,5,10,15,20,30,50,70],
                'min_samples_leaf':[1,2,5,10,15,20], 
                'min_samples_split':[2,5,10,15,20]
                  }
n_iter_search = 10
reg = RandomForestRegressor(verbose=1,n_jobs=-1)
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_search = RandomizedSearchCV(reg, param_distributions=param_dist, n_iter=n_iter_search,
                                   scoring='neg_mean_absolute_error',cv=7) 
random_search.fit(x_train, y_train)
```

```python
random_search.best_estimator_
```

```python
rf=RandomForestRegressor(bootstrap=False, max_depth=50, max_features=30,
                      min_samples_leaf=15, min_samples_split=20, n_jobs=-1,
                      verbose=1)
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
# plot the error difference
error = y_test - y_pred
fig, ax = plt.subplots()
ax.scatter(y_test, error)
ax.axhline(lw=3, color='black')
ax.set_xlabel('Observed')
ax.set_ylabel('Error')
plt.show()
```
